```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import csv
```

# United States Total Cases and Deaths Data

```python
filename = "https://raw.githubusercontent.com/nytimes/covid-19-data/master/live/us-states.csv"

data = pd.read_csv(filename)


df = pd.DataFrame(data, columns = ['date', 'state', 'flips', 'cases', 'deaths'])

df_1 = df.drop(['flips'], axis = 1)
```

```python
print(df_1)
```

# Death Percentage and Survival Rate Calculated and added

```python
df_1['Death_Percent'] = df_1.deaths/df_1.cases * 100
df_1['Survival_Rate'] = 100 - df_1.Death_Percent 
df_1.sort_values(by=['cases'], inplace=True)
print(df_1)
```

```python
df_1.to_csv('us-states.csv')
```

# Total Cases Per State

```python
plt.figure(figsize=(32,32))
plt.barh(df_1['state'], df_1['cases'])
plt.title('Total Cases')
plt.rcParams.update({'font.size': 33})
plt.show()
```

```python
df_2 = df_1

df_2.sort_values(by=['Death_Percent'], inplace=True)
print(df_2)
```

# Death Percentage in the United States

```python

plt.figure(figsize=(32,32))
plt.barh(df_2['state'], df_2['Death_Percent'], color = 'red')
plt.title('Death Percentage')
plt.rcParams.update({'font.size': 33})

plt.show()
```
